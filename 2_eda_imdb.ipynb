### 6. Demuestra que es mejor tener nombres de columnas sin espacios (notación corchetes y notación punto-variable).

```python
cols_con_espacio = [c for c in df.columns if ' ' in c]
if cols_con_espacio:
    orig = cols_con_espacio[0]
else:
    orig = 'Nueva Columna'
    df[orig] = df[df.columns[0]]  

print("Nombre original de la columna seleccionada:", orig)
print("\nAcceso con corchetes (funciona aunque tenga espacios):")
display(df[orig].head())


limpio = orig.lower().replace(" ", "_").replace("-", "_").replace(".", "").replace("á","a").replace("é","e").replace("í","i").replace("ó","o").replace("ú","u")
print(f"\nNombre sugerido limpio: {limpio}")

print("\nIntentando getattr(df, limpio) antes de renombrar (debe fallar porque esa columna aún no existe):")
try:
    display(getattr(df, limpio).head())
except Exception as e:
    print("Error:", type(e).__name__, "-", e)


df = df.rename(columns={orig: limpio})

print("\nColumnas ahora (renombrada):")
print([c for c in df.columns if c == limpio])

print("\nAcceso con corchetes tras renombrar:")
display(df[limpio].head())

print("\nAcceso con notación punto tras renombrar (ahora funciona):")

display(getattr(df, limpio).head())
```

### 7. Renombra las columnas que tengan espacios

```python

import re

def limpiar_nombre(col):
    s = str(col).strip().lower()
    s = s.replace("á","a").replace("é","e").replace("í","i").replace("ó","o").replace("ú","u").replace("ñ","n")
    s = re.sub(r"[ \-\.]+", "_", s)          
    s = re.sub(r"[^\w]", "", s)             
    s = re.sub(r"_+", "_", s).strip("_")     
    return s

mapping = {col: limpiar_nombre(col) for col in df.columns}


seen = {}
final_map = {}
for orig, new in mapping.items():
    if new in seen:
        seen[new] += 1
        new = f"{new}_{seen[new]}"
    else:
        seen[new] = 0
    final_map[orig] = new

df = df.rename(columns=final_map)

print("Columnas renombradas (ejemplo):")
for o, n in list(final_map.items())[:20]:
    if o != n:
        print(f"  '{o}' -> '{n}'")
```

### 10. Inspecciona si hay columnas que tengan valores `NA`

```python

na_count = df.isna().sum()
na_with = na_count[na_count > 0].sort_values(ascending=False)

if na_with.empty:
    print("No hay valores NA en ninguna columna.")
else:
    na_pct = (na_with / len(df) * 100).round(2)
    display(pd.concat([na_with, na_pct], axis=1, keys=['na_count','na_pct']))

    no_na = list(df.columns[df.isna().sum() == 0])
    print(f"\nColumnas sin NA ({len(no_na)}): {no_na}\n")

   
    top_cols = list(na_with.index[:5])
    for col in top_cols:
        print(f"Ejemplo de filas con NA en '{col}':")
        display(df[df[col].isna()].head(5))
```

### 11. Imprime el número total de valores NA que haya en cada columna. Hazlo primero para la columna `Metascore` y después utiliza un ciclo for para hacerlo para todas las columnas

```python

candidatas_metascore = ['Metascore', 'metascore', 'meta score', 'meta_score', 'Meta_Score']
col_metascore = next((c for c in candidatas_metascore if c in df.columns), None)

total = len(df)
if col_metascore is not None:
    na = df[col_metascore].isna().sum()
    pct = round(na / total * 100, 2)
    print(f"Columna encontrada para 'Metascore': '{col_metascore}' -> NA: {na} ({pct}%)")
    if na > 0:
        print(f"Ejemplos de filas con NA en '{col_metascore}':")
        display(df[df[col_metascore].isna()].head(5))
else:
    print("No se encontró una columna llamada 'Metascore' (revisa nombres de columnas).")

print("\n--- NA por todas las columnas ---")
na_counts = df.isna().sum()
na_pct = (na_counts / total * 100).round(2)
res = pd.concat([na_counts, na_pct], axis=1)
res.columns = ['na_count', 'na_pct']
display(res.sort_values('na_count', ascending=False))


for col, row in res.sort_values('na_count', ascending=False).iterrows():
    print(f"{col}: {int(row['na_count'])} NA ({row['na_pct']}%)")
```

### 12. Usa la magia de `dropna()`

```python

print("Shape original:", df.shape)
print("NA por columna (resumen):")
display(df.isna().sum().sort_values(ascending=False).head(10))


df_drop_any = df.dropna()
print("\nDespués dropna() (any) - Shape:", df_drop_any.shape)


candidatas = ['metascore', 'rating']
col_sub = next((c for c in candidatas if c in df.columns), None)
if col_sub:
    df_drop_subset = df.dropna(subset=[col_sub])
    print(f"\nDespués dropna(subset=[{col_sub}]) - Shape:", df_drop_subset.shape)
    print(f"Ejemplos de filas con {col_sub} (sin NA):")
    display(df_drop_subset[[col_sub]].head())
else:
    print("\nNo se encontró 'metascore' ni 'rating' para usar en subset.")


thresh = int(len(df) * 0.7)
df_keep_cols = df.dropna(axis=1, thresh=thresh)
print(f"\nDespués dropna(axis=1, thresh={thresh}) - Shape:", df_keep_cols.shape)
print("Columnas conservadas (ejemplo):", list(df_keep_cols.columns)[:10])


df_clean = df.dropna(subset=[col_sub]) if col_sub else df.copy()
print("\ndf_clean shape (ejemplo):", df_clean.shape)


print("\nNA por columna - antes (top 5):")
display(df.isna().sum().sort_values(ascending=False).head(5))
print("\nNA por columna - después (top 5) en df_clean):")
display(df_clean.isna().sum().sort_values(ascending=False).head(5))
```

### 13. Vuelve a ver la info del dataset

```python
def mostrar_info(df_obj, name="df"):
    print(f"--- Info: {name} ---")
    print("Shape:", df_obj.shape)
    print("Número de columnas:", df_obj.shape[1])
    print("Número de filas:", df_obj.shape[0])
    print("\nInfo del dataframe:")
    df_obj.info()
    print("\nTipos de datos:")
    display(df_obj.dtypes)
    print("\nValores únicos por columna (nunique):")
    display(df_obj.nunique().sort_values(ascending=False).to_frame(name='nunique'))
    na = df_obj.isna().sum()
    na_pct = (na / len(df_obj) * 100).round(2)
    print("\nValores NA por columna (cantidad y %):")
    display(pd.concat([na, na_pct], axis=1, keys=['na_count','na_pct']).sort_values('na_count', ascending=False))
    print("\nEstadísticos descriptivos (transpuesto):")
    display(df_obj.describe(include='all').T)
    print("\nUso de memoria total (bytes):", df_obj.memory_usage(deep=True).sum())
    print("--- fin ---\n")

mostrar_info(df, "df")

if 'df_clean' in globals():
    mostrar_info(df_clean, "df_clean")
```

### 21. Obten los valores únicos de la variable Rating y la frecuencia total de cada uno de estos valores. Posteriormente crea un nuevo dataframe con essos valores
